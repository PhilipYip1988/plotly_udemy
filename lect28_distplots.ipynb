# Import Data Science Libraries

To get started, we need to import the data science libraries. We will use numpy to create data or pandas to read in data from a csv file:

```python
import numpy as np
import numpy.random as random
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.figure_factory as ff
random.seed(101)
```

We will create 1000 randomly normalised datapoints:

```python
x = random.randn(1000)
```

# FigureFactory

The figure factory contains a number of specialised plots, that often include subplots. 


# Distplot

The distplot for example displays a histogram plot and kernel density estimation (kde) in the top subplot, accompanied by a rugplot in the bottom subplot. We require a hist_data and group_labels:

```python
hist_data = [x]
group_labels = ["distplot"]
```

We can create a simple histplot:

```python
fig = ff.create_distplot(hist_data, group_labels)
pyo.iplot(fig)
```

To update the properies of the figure such as the height of the layout, we can use the fig method update_figure:

```python
fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(height=600)
pyo.iplot(fig)
```

Let's now create hist_data which contains four arrays of data:

```python
x1 = random.randn(200)-2
x2 = random.randn(200)
x3 = random.randn(200)+2
x4 = random.randn(200)+4
```

```python
hist_data = [x1, x2, x3, x4]
group_labels = ["x1", "x2", "x3", "x4"]
```

```python
fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(height=600)
pyo.iplot(fig)
```

We can modify the bin_size for each dataset:

```python
fig = ff.create_distplot(hist_data, group_labels, bin_size=[2, 1, 3, 4])
fig.update_layout(height=600)
pyo.iplot(fig)
```

```python
fig = ff.create_distplot(hist_data, group_labels, bin_size=[.2, .1, .3, .4])
fig.update_layout(height=600)
pyo.iplot(fig)
```

We can look at the relatively frequency of three letter words in the Snodgrass and Twain writings:

```python
snodgrass = [.209, .205, .196, .210, .202, .207, .224, .223, .220, .201]
twain = [.225, .262, .217, .240, .230, .229, .235, .217]
```

```python
hist_data = [snodgrass, twain]
group_labels = ["Snodgrass Writings", "Mark Twain Writings"]
```

```python
fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(height=600)
pyo.iplot(fig)
```

We need to update the bin_size in order to view this dataset:

```python
fig = ff.create_distplot(hist_data, group_labels, bin_size=[0.005, 0.005])
fig.update_layout(height=600)
pyo.iplot(fig)
```
