# Import Data Science Libraries

To get started, we need to import the data science libraries. We will use numpy to create data or pandas to read in data from a csv file:

```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

# Import Data

We will import a dataset which gives some measurements of marine snails and examine its head:

```python
df = pd.read_csv("./data/abalone.csv")
df.head()
```

# Histogram

```python
data = [go.Histogram(x=df["length"])]
layout = go.Layout(title="Histogram",
                   xaxis=dict(title="length"), 
                   yaxis=dict(title="count"),
                   height=600)
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```

Specifying the range and size of the xbins

```python
data = [go.Histogram(x=df["length"], xbins=dict(start=0, end=1, size=0.02))]
layout = go.Layout(title="Histogram",
                   xaxis=dict(title="length"), 
                   yaxis=dict(title="count"),
                   height=600)
fig = go.Figure(data=data, layout=layout)
pyo.iplot(fig)
```
