# Import Data Science Libraries

To get started, we need to import the data science libraries. We will use numpy to create data or pandas to read in data from a csv file:

````python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
import plotly.figure_factory as ff
````

# Import Data

We will import the iris data as ```df``` and examine the head:

````python
df = pd.read_csv("./data/iris.csv")
df.head()
````

Let's have a look at the unique values in the series cass:

````python
df["class"].unique()
````

# Distplot

We can use list comprehension to create the list hist_data of each clases petal length:

````python
hist_data = [df[df["class"] == species]["petal_length"] for species in df["class"].unique()]
group_labels = df["class"].unique()
````

We can then create a distplot of these values:

````python
fig = ff.create_distplot(hist_data, group_labels)
fig.update_layout(height=600, xaxis=dict(title="petal length"))
pyo.iplot(fig)
````
